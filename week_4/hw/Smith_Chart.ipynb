```python
import numpy as np
from matplotlib import pyplot as plt
```

> Function

```python
def SmithChart(ax,Smith='Z'):

    # set parameters
    ax.set_aspect('equal', 'datalim') 
    ax.spines["right"].set_color("none")  
    ax.spines["top"].set_color("none")  
    ax.spines["left"].set_color("none")  
    ax.spines["bottom"].set_color("none")  
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.tick_params(color='none') 
    # Smith chart　Drawing
    #(1) Circle
    t=np.linspace(0,2*np.pi,100) 
    x=np.cos(t)
    y=np.sin(t)
    ax.plot(x,y,color='black',linewidth=2)
    #(2) Line 
    x0=np.linspace(-1,1,10) 
    y0=np.zeros(10)
    ax.plot(x0,y0,'k',linewidth=1)
    #(3) R G Circle　
    k=[.25,.5,.75] 
    for i in range(0,len(k)):
        x1=k[i]+(1 - k[i])*np.cos(t)# offset + radious*cos(angle)  (.25,.75),(.5,.5),(.75,.25)
        y1=(1 - k[i]) * np.sin(t) # sin(angle)
        if(Smith=='Z'):
            ax.plot(x1,y1,'k',lw=0.5)
        if(Smith=='Y'):
            ax.plot(-x1,-y1,'r',lw=0.5)
    # (4) X,B Arc 
    kt=[2.5,np.pi,3.79,4.22] #　4 lines
    k2 = [.5,1,2,4]
    for i in range(0,len(kt)):
        t = np.linspace(kt[i], 1.5*np.pi, 50) # start=kt[i],stop~1.5π   split into 50 parts
        a = 1 + k2[i] * np.cos(t) # offset(=1) + radious*cos(angle)
        b = k2[i] + k2[i] * np.sin(t) #  
        if(Smith=='Z'):
            ax.plot(a, b,'k',lw=0.5) # y +reigion
            ax.plot(a, -b,'k',lw=0.5)# y -reigion
        if(Smith=='Y'):
            ax.plot(-a, b,'r:',lw=0.5)
            ax.plot(-a, -b,'r:',lw=0.5)
            
    return 0
```

> Main

```python
fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(1,3,1)
SmithChart(ax1)

ax2 = fig.add_subplot(1,3,2)
SmithChart(ax2,Smith='Y')

ax3 = fig.add_subplot(1,3,3)
SmithChart(ax3)
SmithChart(ax3,Smith='Y')
```
